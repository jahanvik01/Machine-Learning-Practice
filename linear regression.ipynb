```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# importing csv file

df = pd.read_csv("canada_per_capita_income.csv")
df.head()
```

```python
# to check relation between 'year' and 'per capita income'

plt.xlabel('year')
plt.ylabel('per capita income')
plt.scatter(df.year, df['PCI'], color='g')
plt.show()
```

```python
# providing data to model
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df.PCI)
```

```python
# predicting PCI
reg.predict([[2020]])
```

```python
# coefficient
reg.coef_
```

```python
# intercept
reg.intercept_
```
